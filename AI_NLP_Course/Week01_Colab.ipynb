```python
# ==============================
# Week 1 - Google Colab & Python Basics
# NumPy, Matplotlib, Scikit-learn, TensorFlow
# ==============================
# TIP: If you encounter an error, click the "Explain error" button
#      or ask Colab's AI assistant (Ctrl+Shift+Space) for help.

# ==============================
# 1. Google Colab Basics
# ==============================
# Ctrl+Enter  : Run current cell
# Shift+Enter : Run and move to next cell
# +Code       : Add new code cell
# +Text       : Add markdown cell
# Ctrl+M -    : Split cell at cursor
# Ctrl+M M    : Merge with next cell

print("Hello, Colab!")
print(f"Python is ready.")


# ==============================
# 2. NumPy — Array Operations
# ==============================
import numpy as np

a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print("=" * 40)
print("2. NumPy Basics")
print("=" * 40)
print(f"a = {a}")
print(f"b = {b}")
print(f"a + b = {a + b}")
print(f"a * b = {a * b}")
print(f"Mean of a: {a.mean():.2f}")
print(f"Std of a:  {a.std():.2f}")

matrix = np.random.randn(3, 4)
print(f"\nRandom Matrix (3x4):\n{matrix}")
print(f"Shape: {matrix.shape}")

# ==============================
# 3. Matplotlib — Visualization
# ==============================
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(x, np.sin(x), 'r-', label='sin(x)')
ax1.plot(x, np.cos(x), 'b-', label='cos(x)')
ax1.set(title='Line Plot', xlabel='x', ylabel='y')
ax1.legend(); ax1.grid(True, alpha=0.3)

data = np.random.randn(200)
ax2.hist(data, bins=20, color='#6c5ce7', edgecolor='white')
ax2.set(title='Histogram (Normal Distribution)', xlabel='Value', ylabel='Count')
ax2.grid(True, alpha=0.3)

fig.suptitle('Matplotlib Basics', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout(); plt.show()

# ==============================
# 4. Scikit-learn — Simple Classification
# ==============================
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

print("=" * 40)
print("4. Scikit-learn Basics")
print("=" * 40)

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.3, random_state=42)

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(f"Dataset: IRIS ({iris.data.shape})")
print(f"Classes: {list(iris.target_names)}")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")

# ==============================
# 5. TensorFlow — Simple Neural Network
# ==============================
import tensorflow as tf

print("\n" + "=" * 40)
print("5. TensorFlow Basics")
print("=" * 40)
print(f"TensorFlow version: {tf.__version__}")

model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50, verbose=0)
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Keras MLP Accuracy: {acc:.4f}")

# --- sklearn vs keras comparison ---
fig, ax = plt.subplots(figsize=(6, 4))
names = ['sklearn\nLogistic', 'Keras\nMLP']
accs = [accuracy_score(y_test, y_pred), acc]
bars = ax.bar(names, accs, color=['#0984e3', '#e74c3c'], edgecolor='white', width=0.4)
for b, v in zip(bars, accs):
    ax.text(b.get_x() + b.get_width()/2, v + 0.005, f'{v:.4f}',
            ha='center', fontweight='bold', fontsize=13)
ax.set(title='IRIS Classification — sklearn vs Keras', ylabel='Test Accuracy')
ax.set_ylim(min(accs) - 0.05, max(accs) + 0.04)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout(); plt.show()

# ==============================
# TROUBLESHOOTING GUIDE
# ==============================
# If you see an error:
#   1. Click "Explain error" button below the error message
#   2. Use Colab AI: Edit > AI > Ask AI (or Ctrl+Shift+Space)
#   3. Copy the error message and ask the AI chatbot
#
# Common issues:
#   - ModuleNotFoundError  → !pip install <package>
#   - Shape mismatch       → Check array dimensions with .shape
#   - GPU not available    → Runtime > Change runtime type > GPU
```
